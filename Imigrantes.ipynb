# Dados de imigrantes no Brasil


## Coletando os dados 

```python
import pandas as pd
```

```python
df = pd.read_csv("dados/imigrantes_canada.csv")
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True) # Agora o país é o index, podemos localizar o país que quisermos
```

```python
anos = list(map(str, range(1980, 2014))) #colocando os anos na ordem
```

```python
anos
```

## Coletando os dados do Brasil

```python
brasil = df.loc['Brasil', anos] #localizando o brasil no df
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()} #Criando um dicionario com ano e n° imigrantes

dados_brasil = pd.DataFrame(brasil_dict) # colocando o dicionario em data frame  
```

```python
dados_brasil.head
```

## Usando matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(8, 4)) #tamanho da figura
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) # plotando os dados
plt.title('Imigração do Brasil para o Canadá') #titulo
plt.xlabel('Ano') # eixo X
plt.ylabel('Número de imigrantes') #eixo Y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) # intervalo de tempo do eixo X
plt.show() # Para não mostar números indesejados encima do gráfico
```

## Criando subplots em duas direções

```python
fig, axs = plt.subplots(2,2, figsize=(10,6)) # Colocando quantos subplots existirão, no caso, 2 na linha de cima e 2 na de baixo
fig.subplots_adjust(hspace=0.5, wspace=0.3) # Ajustando o espaçamento
fig.suptitle('Imigração dos 4 maiores países da América do Sul para o Canadá de 1980 a 2013') # Titulo geral

axs[0,0].plot(df.loc['Brasil', anos]) # localizando o subplot do Brasil
axs[0,0].set_title('Brasil') 

axs[0,1].plot(df.loc['Colômbia', anos]) # localizando o subplot da Colômbia
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos]) # localizando o subplot da Argentina
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos]) # localizando o subplot do Peru
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # Colocando os anos de 5 em 5 para todos os gráficos

for ax in axs.flat:
    ax.set_xlabel('Ano') # Ano para todos
    ax.set_ylabel('Número de Imigrantes') # Número de imigrantes para todos
    
    
ymin = 0 # Linha y minima
ymax = 7000 # Linha y maxima

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax) # Colocando em todos os gráficos, o limite, de 0 a 7000 
    
plt.show()
```

## Atividade III

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}



dados = pd.DataFrame(vendas_2022, index=lojas)
dados
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6)) 
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Panorama anual das 4 lojas')

axs[0,0].plot(dados.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(dados.loc['B'])
axs[0,1].set_title('Loja B') 

axs[1,0].plot(dados.loc['C'])
axs[1,0].set_title('Loja C') 

axs[1,1].plot(dados.loc['D'])
axs[1,1].set_title('Loja D') 

for ax in axs.flat:
    ax.set_xlabel('Mês') 
    ax.set_ylabel('Número de Vendas') 
    
ymin = 0 
ymax = 500 

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax) 


plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)

cores = []
for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 à 2013', loc='left', fontsize=18, color='black')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

plt.show()
```
